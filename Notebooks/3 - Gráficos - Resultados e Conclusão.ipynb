# Visualizações

```python
sns.set_context('paper', font_scale=2.4)
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

## Acurácia Método 1

```python
plt.figure(figsize=(20,10),frameon=False)
plt.title('Acurácia de Classificadores - Características Históricas')
plt.ylabel('Acurácia')
plt.xlabel('Características')
plot = sns.barplot(x=algoritmos,y=m1acurácias,color='blue')

fig = plot.get_figure()

fig.savefig('Acurácia Método1.png')
```

## Acurácia Método 2

```python
plt.figure(figsize=(20,10),frameon=False)
plt.title('Acurácia de Classificadores - Características RFE')
plt.ylabel('Acurácia')
plt.xlabel('Características')
plot = sns.barplot(x=algoritmos,y=m2acurácias,color='blue')

fig = plot.get_figure()

fig.savefig('Acurácia Método2.png')
```

## Comparação de acurácias

```python
plt.subplots(figsize=(18,10))
ind = np.arange(len(algoritmos))
p1 = plt.bar(ind, m1n_caracteristicas,0.35)
p2 = plt.bar(ind + 0.35, m2n_caracteristicas,0.35)

plt.ylabel('Características')
plt.title('Nº de Características por Classificador')
plt.xticks(ind + 0.15, algoritmos)
# plt.yticks(np.arange(0,0.1,1))
plt.legend((p1[0], p2[0]), ('Características históricas', 'Características RFE'))


plt.savefig('ComparativoCaracs.png', bbox_inches='tight')
```

```python
plt.subplots(figsize=(18,10))
ind = np.arange(len(algoritmos))
p1 = plt.bar(ind, m1acurácias,0.35)
p2 = plt.bar(ind + 0.35, m2acurácias,0.35)

plt.ylabel('Acurácia')
plt.title('Acurácia por Classificador')
plt.xticks(ind + 0.15, algoritmos)
# plt.yticks(np.arange(0,0.1,1))
plt.legend((p1[0], p2[0]), ('Características históricas', 'Características RFE'), bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)


plt.savefig('ComparativoAcuracias.png', bbox_inches='tight')
```
